4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9],
obtenga la media aritmética de la diferencia (a-b).

```python
# Definir los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Se calcula la diferencia entre los elementos de a y b
diferencias = a - b

# Calcular la media aritmética de las diferencias
mediadiferencias = np.mean(diferencias)

print("La media aritmética de la diferencia entre los elementos de a y b es:", mediadiferencias)
```

5) Tengo valores de cordenadas (x, y) en las columnas
del array
1.33 4.5
30.0 10.7
70.2 0.5
Agregue a este array las coordenadas (37.1, -3.6).
Muestre en pantalla las dimensiones del nuevo
array.

```python
# Array original
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

# Coordenadas a agregar
nuevacoordenada = np.array([37.1, -3.6])

# Agregamos las nuevas coordenadas al array original
coordenadas_connueva = np.vstack([coordenadas, nuevacoordenada])

# Mostramos las dimensiones del nuevo array
print("Dimensiones del nuevo array:", coordenadas_connueva.shape)
```

6) Copie el array del ejercicio anterior. Traspóngalo (.
Agregue ahora dos nuevos pares de coordenadas: (10.8,
3.0) y (35.8, 12.0).

```python
# Creamos un array con coordenadas x e y
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

# Imprimimos las coordenadas originales
print("Coordenadas originales:")
print(coordenadas)

# Agregamos dos nuevas coordenadas al array original
coordenadas = np.append(coordenadas, [[37.1, -3.6]], axis=0)

# Imprimimos las dimensiones del nuevo array
print("\nDimensiones del nuevo array después de agregar una coordenada:")
print(coordenadas.shape)

# Transponemos el array
coordenadas_transpuesta = np.transpose(coordenadas)

# Imprimimos el array transpuesto
print("\nArray transpuesto:")
print(coordenadas_transpuesta)

# Agregamos dos nuevos pares de coordenadas al array transpuesto
coordenadas_transpuesta = np.append(coordenadas_transpuesta, [[10.8, 3.0], [35.8, 12.0]], axis=1)

# Imprimimos las dimensiones del nuevo array transpuesto
print("\nDimensiones del nuevo array transpuesto después de agregar dos coordenadas:")
print(coordenadas_transpuesta.shape)
```

1) ¿Cuántos Municipios tienen más de 100000

```python
import random

# Leemos el archivo CSV y lo carga en un DataFrame
df = pd.read_csv('NucleosPoblacion.csv')

# Filtramos los municipios con más de 100000 habitantes
municipiosmas100k = df[df['Poblacion'] > 100000]

# Contamos los municipios que que cumplen la condición
numero_municipios_mas_100k = municipiosmas100k.shape[0]

print("Número de municipios con más de 100,000 habitantes:", numero_municipios_mas_100k)
```

2) Realice una gráfica de barras sobre la polbación de cada
ciudad, ordenela de menor a mayo y responda:

```python
import matplotlib.pyplot as plt

df = pd.read_csv('NucleosPoblacion.csv')

# Ordenamos el DataFrame por la población de menor a mayor
df_sorted = df.sort_values(by='Poblacion')

# Creamos la gráfica de barras
plt.figure(figsize=(10, 6))
plt.barh(df_sorted['Municipio'], df_sorted['Poblacion'], color='skyblue')
plt.xlabel('Población')
plt.ylabel('Ciudad')
plt.title('Población de cada ciudad')
plt.grid(axis='x')

# Mostramos la gráfica
plt.show()

# Respuesta a la pregunta a
segunda_ciudad_mas_poblada = df_sorted.iloc[-2]['Municipio']
print("a. La segunda ciudad más poblada es:", segunda_ciudad_mas_poblada)

# Respuesta a la pregunta b
posicion_granada = df_sorted[df_sorted['Municipio'] == 'Granada'].index[0] + 1
print("b. Granada ocupa la posición", posicion_granada, "en el ranking de las ciudades más pobladas.")
```

4) ¿Cuál es el municipio situado más al Norte? (Usar el
valor de la coordenada "Y" que representa la latitud en
grados). Proporcione también la provincia a la que
pertenece y su población.

```python
df = pd.read_csv('NucleosPoblacion.csv')

# Encuentra el índice del municipio situado más al norte
indiceMunicipioMaNorte = df['Y'].idxmax()

# Extrae la información del municipio más al norte
nombreMunicipio_masNorte = df.at[indiceMunicipioMaNorte, 'Municipio']
provincia_municipio_mas_norte = df.at[indiceMunicipioMaNorte, 'Provincia']
poblacion_municipio_mas_norte = df.at[indiceMunicipioMaNorte, 'Poblacion']

print("El municipio situado más al norte es:", nombreMunicipio_masNorte)
print("Provincia:", provincia_municipio_mas_norte)
print("Población:", poblacion_municipio_mas_norte)
```

5) Obtenga la media, mediana, desviación estándar, valor
máximo y valor mínimo de la población de los municipios
de la provincia de Granada.

```python
df = pd.read_csv('NucleosPoblacion.csv')

# Filtramos los datos para la provincia de Granada
datosGranada = df[df['Provincia'] == 'Granada']

# Obtenemos la población de los municipios de Granada
poblacionGranada = datosGranada['Poblacion']

# Calculamos la media
media = poblacionGranada.mean()

# Calculamos la mediana
mediana = poblacionGranada.median()

# Calculamos la desviación estándar
desviacionEstandar = poblacionGranada.std()

# Obtener el valor máximo
valorMaximo = poblacionGranada.max()

# Obtenemos el valor mínimo
valorMinimo = poblacionGranada.min()

# Imprimir los resultados
print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacionEstandar)
print("Valor máximo:", valorMinimo)
print("Valor mínimo:", valorMaximo)
```

6) Realice un histograma con la población de los
Municipios para cada una de las provincias.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('NucleosPoblacion.csv')

# Agrupamos los datos por provincia
gruposProvincias = df.groupby('Provincia')

# Iteramos sobre cada grupo (provincia) y creamos un histograma de la población de los Municipios de esa provincia
for provincia, datosProvincia in gruposProvincias:
    plt.figure(figsize=(8, 6))
    plt.hist(datosProvincia['Poblacion'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histograma de población de municipios en {provincia}')
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```

7) Seleccione al azar cincuenta municipios diferentes de
entre los diponibles en el archivo. Asegúrese de que no
se repitan. ¿Luego calcule el promedio de la población y
la desviación estándar de esto 50 municipios?.

```python


data = pd.read_csv('NucleosPoblacion.csv')

# Conseguimos una lista de índices únicos de los municipios
indices_unicos = data.index.tolist()

# Seleccionamos al azar cincuenta municipios diferentes
indices_seleccionados = random.sample(indices_unicos, 50)

# Creamos un DataFrame con los municipios seleccionados
municipios_seleccionados = data.loc[indices_seleccionados]

# Calculamos el promedio de la población y la desviación estándar
promedio_poblacion = municipios_seleccionados['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_seleccionados['Poblacion'].std()

print("Promedio de la población:", promedio_poblacion)
print("Desviación estándar de la población:", desviacion_estandar_poblacion)
```
